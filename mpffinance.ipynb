# 1.0. PLOTANDO GRÁFICOS

```python
#Plotando um gráfico de barra
mpf.plot(df, style='mike')
```

```python
#Plotando um gráfico de candle
mpf.plot(df, type='candle', style=('mike'))
```

```python
#Plotando um gráfico de linha
mpf.plot(df, type='line', style=('mike'))
```

```python
#Plotando um gráfico de renko
mpf.plot(df, type='renko', style=('mike'))
```

```python
#Plotando um gráfico de ponto & figura 
mpf.plot(df, type='pnf', style=('mike'))
```

## 1.2. médias móveis

```python
mpf.plot(df, type='ohlc', mav = 20 , style=('mike'))
```

```python
mpf.plot(df, type='candle', mav = (20, 50, 200) , style=('mike'))
```

## 1.3. Volume financeiro 

```python
mpf.plot(df, type='ohlc', mav = 20 , volume = True, style=('mike'))
```

```python
mpf.plot(df, type='candle', mav = (20, 50, 200) , volume=True, style=('mike'))
```
